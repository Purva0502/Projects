```python
#Import Libraries 
import pandas as pd
import numpy as np
import os 
import pickle
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report     
```

```python
#Load Data Set
df=pd.read_csv('parkinsons.data')
display (df)
```

```python
#Create X

X=df.drop(labels=['status','name'],axis=1)
display (X.head())

#Create Y
Y=df['status']
display (Y.head())
```

```python
#Load Pickle File and Prediction with Pickle File 
model=pickle.load(open('deploy_SVM.pkl','rb'))
# Prediction 
prediction = model.predict (X)
print ( 'Accuracy ',accuracy_score (Y, prediction))
print ()
print ( 'Classification Report')
print (classification_report  (Y,prediction,zero_division=0))
print ( 'Confusion Matrix')
print (confusion_matrix  (Y,prediction))
```

```python
#Display the Result 
result = pd.concat ([df['status'],pd.DataFrame (prediction,columns = ['Prediction'])],axis =1)
display (result.head(50))
```
